```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

__Read stock info for AMZN__

```python
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 2, 15)
```

```python
amzn = data.DataReader("AMZN", 'yahoo', start, end)
```

```python
myamzn = amzn.reset_index()
```

__Plot a line chart of volume from 1/1/2022 until today using ggplot style__

```python
plt.style.use('ggplot')
myamzn.plot.line(x='Date', y='Volume', figsize=(12,5), lw=3, color='teal')
```

__Plot a bar chart of the opening for the month of December 2021__

```python
decStart = datetime.datetime(2021, 12, 1)
decEnd = datetime.datetime(2021, 12, 31)
decamzn = data.DataReader("AMZN", 'yahoo', decStart, decEnd)
```

```python
newamzn = decamzn.reset_index()
```

```python
newamzn.plot.bar(x='Date',y='Open', figsize=(18,5), color='lightblue')
```
